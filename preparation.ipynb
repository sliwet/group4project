```python
import json
import pandas as pd
from pprint import pprint
```

```python
with open('Resources/city.list.json', 'r',encoding='utf-8') as jsonfile:
    jsndata = json.load(jsonfile)

ids = []
cit = []
lat = []
lon = []

for i in range(len(jsndata)):
    if jsndata[i]['country'] == 'US':
        ids.append(jsndata[i]['id'])
        cit.append(jsndata[i]['name'])
        lat.append(jsndata[i]['coord']['lat'])
        lon.append(jsndata[i]['coord']['lon'])

dftemp = pd.DataFrame({'ID':ids,'City':cit,'Latitude':lat,'Longitude':lon})
dftemp.to_csv("Resources/uscities-all.csv", index=False, header=True)
```

```python
uscities_all = pd.read_csv("Resources/uscities-all.csv", encoding="utf-8")
print(f"# of US cities: {len(uscities_all)}")
uscities_all.head()
```

```python
uscities_major = pd.read_csv("Resources/uscities-major.csv", encoding="utf-8")
print(f"# of major US cities: {len(uscities_major)}")
uscities_major.head()
```

```python
citylist = uscities_all['City'].tolist()
ids = []
lat = []
lon = []
for city in uscities_major['City']:
    try:
        icity = citylist.index(city)
        ids.append(uscities_all['ID'][icity])
        lat.append(uscities_all['Latitude'][icity])
        lon.append(uscities_all['Longitude'][icity])
    except:
        ids.append("")
        lat.append("")
        lon.append("")
uscities_major['ID'] = ids
uscities_major['Latitude'] = lat
uscities_major['Longitude'] = lon
    
```
